```python
#droping the columns
df = df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
#drop the duplicated in the data set
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].unique()
```

```python
df.rate.isnull().sum()
```

```python
# fill the null values with mean
df['rate'].fillna(df['rate'].mean(),inplace = True)
```

```python
df.rate.isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.info() # no null values
```

```python
# renaming the columns
df.rename(columns = {'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# will keep the one column
df = df.drop(['listed_in(city)'],axis=1)
```

```python
df.head()
```

```python
df['Cost2plates'].unique()
```

```python
# above we can see there are comma's in the rate

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
    
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

```python
df.head()
```

```python
# rest type column

df.rest_type.value_counts()
```

```python
# we can replace the columns which contain the less count to others
rest_types = df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types < 1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df.location.value_counts()
```

```python
location = df['location'].value_counts(ascending=False)
location_lessthan300 = location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
    
df['location'] = df['location'].apply(handle_location)
df.location.value_counts()
```

```python
df.head()
```

```python
#clean the cuisines
df.cuisines.value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending=False)

cuisin_lessthan100 = cuisines[cuisines<100]

def handle_cusines(value):
    if(value in cuisin_lessthan100):
        return 'others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(handle_cusines)
df['cuisines'].value_counts()
```
